```python
import pandas as pd   
import matplotlib.pyplot as plt 
import csv
import re
import numpy as np
from collections import Counter
from matplotlib.pyplot import figure
import statistics
```

```python
import numpy as np
def _sum(arr): 
    sum=0
    for i in arr:
        sum = sum + i         
    return(sum) 

def _max(arr):
    max = 0
    max = np.max(arr)   
    return(max)

def _min(arr):
    min = 0
    min = np.min(arr)   
    return(min)


```

# newsau



```python
newsau = pd.read_csv("newsau_Analysis.csv")
newsau = newsau[newsau.News != "[]"]
newsau = newsau[newsau.Subjectivity >0]

Newsau_std = statistics.stdev(newsau.Subjectivity)

newsau
```

```python
#find a number of Positive and Negative
word_counts = Counter(newsau.Analysis)
newsau_Positive = word_counts['Positive']
newsau_Negative = word_counts['Negative']

totalsAnalysis= newsau_Positive + newsau_Negative


#show number of Positive and Negative

#print(newsau_Positive)
#print(newsau_Negative)

#find a percentage of Positive and Negative
newsau_Positive_P = '{percent:.1%}'.format(percent=newsau_Positive/totalsAnalysis)
newsau_Negative_P = '{percent:.1%}'.format(percent=newsau_Negative/totalsAnalysis)


print('The percentage of newsau Positive =' + newsau_Positive_P)
print('The percentage of newsau Negative =' + newsau_Negative_P)
```

```python
#Average of the newsau_Subjectivity

arr=(newsau.Subjectivity)
newsau_Subjectivity__Average = _sum(arr) 
newsau_Subjectivity__Average= newsau_Subjectivity__Average/len(newsau.Subjectivity)
print ('Average of the newsau_Subjectivity is ', newsau_Subjectivity__Average) 

#maximum value of the newsau

arr=(newsau.Subjectivity)
newsau_Subjectivity__max = _max(arr) 
print ('The newsau_Subjectivity maximum value is', newsau_Subjectivity__max) 

#minimum value of the newsau
arr=(newsau.Subjectivity)
newsau_Subjectivity__min = _min(arr) 
print ('The newsau_Subjectivity minimum value is', newsau_Subjectivity__min) 

#Average of the newsau_Polarity

arr=(newsau.Polarity)
newsau_Polarity__Average = _sum(arr) 
newsau_Polarity__Average= newsau_Polarity__Average/len(newsau.Polarity)
print ('Average of the newsau_Polarity is ', newsau_Polarity__Average) 

#maximum value of the newsau

arr=(newsau.Polarity)
newsau_Polarity__max = _max(arr) 
print ('The newsau_Polarity maximum value is', newsau_Polarity__max) 

#minimum value of the newsau
arr=(newsau.Polarity)
newsau_Polarity__min = _min(arr) 
print ('The newsau_Polarity minimum value is', newsau_Polarity__min) 
```
