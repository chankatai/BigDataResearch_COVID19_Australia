```python
import string
from collections import Counter
import matplotlib.pyplot as plt
```

# newsau

```python
text = open('newsau_output.txt',encoding='utf-8').read()
lower_case = text.lower()
cleaned_text = lower_case.translate(str.maketrans('', '', string.punctuation))

tokenized_words = cleaned_text.split()
stop_words = ["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]

final_words = []
for word in tokenized_words:
    if word not in stop_words:
        final_words.append(word)
        
emotion_list = []
with open('emotions.txt','r') as file:
    for line in file:
        clear_line = line.replace("\n", '').replace(",", '').replace("'", '').strip()
        word, emotion = clear_line.split(':')

        if word in final_words:
            emotion_list.append(emotion)

w = Counter(emotion_list)
print(w)

happy = w[' happy']
sad = w[' sad']
fearful = w[' fearful']
attracted = w[' attracted']
powerless = w[' powerless']
attached = w[' attached']
adequate = w[' adequate']
entitled = w[' entitled']
fearless = w[' fearless']
esteemed = w[' esteemed']
anxious = w[' anxious']
cheated = w[' cheated']
singledout = w[' singled out']
average = w[' average']
surprise = w[' surprise']
focused = w[' focused']
free = w[' free']
alone = w[' alone']
loved = w[' loved']
lustful = w[' lustful']
ecstatic = w[' ecstatic']
lost = w[' lost']
codependent = w[' codependent']
derailed = w[' derailed']
safe = w[' safe']
apathetic = w[' apathetic']
obsessed = w[' obsessed']
demoralized = w[' demoralized']
burdened = w[' burdened']
independent = w[' independent']


len(w.keys())
```

```python
plt.figure(figsize=(12,14))
age = 'happy','sad','fearful','attracted','powerless','attached','adequate','entitled','fearless','esteemed','anxious','cheated','singledout','average','surprise','focused','free','alone','loved','lustful','ecstatic','lost','codependent','derailed','safe','apathetic','obsessed','demoralized','burdened','independent'
counts = [happy,sad,fearful,attracted,powerless,attached,adequate,entitled,fearless,esteemed,anxious,cheated,singledout,average,surprise,focused,free,alone,loved,lustful,ecstatic,lost,codependent,derailed,safe,apathetic,obsessed,demoralized,burdened,independent]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lavender','pink']
plt.pie(counts, labels = age, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.axis('equal')
plt.savefig('../FYP_web/static/images/Newsau_SentimentAnalysisPi.jpg')
plt.show()
```
