```python
import requests
import json
import pymongo
import pandas as pd
```

## New South Wales Datasets

```python
# NSW COVID-19 Infection case (JSON) to MongoDB

r = requests.get('https://data.nsw.gov.au/data/datastore/dump/2776dbb8-f807-4fb2-b1ed-184a6fc2c8aa?format=json')
data = r.json()

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["FYP_Test"]

if "Nsw_infection_case" in mydb.list_collection_names():
    mydb["Nsw_infection_case"].drop()

mycol = mydb["Nsw_infection_case"]

fields = [d['id'] for d in data["fields"]]
records = data['records']

all_data = list(map(lambda x: dict(zip(fields, x)), records))

x = mycol.insert_many(all_data) 
```

```python
# NSW COVID-19 clinics (CSV) to MongoDB

data = pd.read_csv('https://data.nsw.gov.au/data/datastore/dump/85da884f-a9f5-4cb3-95e8-d6b81b0d2e3a?bom=True')
data

df = pd.DataFrame(
    data,
    columns =['Clinic Address (field_clinic_address:address_line1)' , 'Clinic Address (field_clinic_address:address_line2)', 'Suburb', 'States', 'Postcode' , 'Clinic Phone', 'Clinic Website', 'Clinic Facility Instruction', 'Clinic Drive Through Testing', 'Clinic Appointment Required', 'Clinic Patient Restriction', 'Clinic Prescription Required']
)
df

out = df.to_json(orient='records')#[1:-1].replace('},{', '} {')

[list(json.loads(out)[i].values()) for i in range(len(json.loads(out)))]

list(df.columns)

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["FYP_Test"]

if "Nsw_COVID-19_clinics" in mydb.list_collection_names():
    mydb["Nsw_COVID-19_clinics"].drop()

mycol = mydb["Nsw_COVID-19_clinics"]

fields = list(df.columns)
records = [list(json.loads(out)[i].values()) for i in range(len(json.loads(out)))]

all_data = list(map(lambda x: dict(zip(fields, x)), records))

x = mycol.insert_many(all_data)
```

## Public Datasets

```python
#Airport Traffic (CSV) to MongoDB

data = pd.read_csv('https://data.gov.au/data/dataset/cc5d888f-5850-47f3-815d-08289b22f5a8/resource/38bdc971-cb22-4894-b19a-814afc4e8164/download/mon_pax_web.csv')
data

df = pd.DataFrame(
    data,
    columns =['AIRPORT' , 'Year', 'Month', 'Dom_Pax_In', 'Dom_Pax_Out' , 'Dom_Pax_Total', 'Int_Pax_In', 'Int_Pax_Out', 'Int_Pax_Total', 'Pax_In', 'Pax_Out', 'Pax_Total']
)
indexNames = df[df['Year'] < 2019].index
df.drop(indexNames, inplace=True)
df.reset_index(drop=True, inplace=True)
df

ids = df['AIRPORT']
years = df['Year']
months = df['Month']

df1 = pd.DataFrame(zip(ids, years, months), columns=["id", "year", "month"])

df["date"] = df1['year'].map(str) + '/' + df1['month'].map(str) + '/' + '28'
df

out = df.to_json(orient='records')#[1:-1].replace('},{', '} {')

[list(json.loads(out)[i].values()) for i in range(len(json.loads(out)))]

list(df.columns)

myclient = pymongo.MongoClient("mongodb://localhost:27017/")
mydb = myclient["FYP_Test"]

if "Airport_Traffic" in mydb.list_collection_names():
    mydb["Airport_Traffic"].drop()

mycol = mydb["Airport_Traffic"]

fields = list(df.columns)
records = [list(json.loads(out)[i].values()) for i in range(len(json.loads(out)))]

all_data = list(map(lambda x: dict(zip(fields, x)), records))

x = mycol.insert_many(all_data)
```
